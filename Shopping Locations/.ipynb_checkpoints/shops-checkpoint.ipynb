```python
import pandas as pd
```

```python
file = "shop_point.csv"
data = pd.read_csv(file)
atlanta_poi = data[['X','Y','shop','name','addr:postcode']]
atlanta_poi['shop'].unique()
```

```python
#Subsets of Atlanta Points of Interest
#1: alc_shops - includes both wine and alcohol shops 
#2: food_shops - bakery, confectionery, deli, frozen yogurt, Frozen Yogurt, ice_cream, fresh juices & smoothies
#3: shopping - accesories, cosmetics, department_store, mall, skateshop, electronics, craft, gift, glam shop, household, interior_decoration
#4: 
```

```python
alc_shops = atlanta_poi.loc[(atlanta_poi['shop']=='alcohol')|(atlanta_poi['shop']=='wine')| (atlanta_poi['shop']=='beverages')]
alc_shops
```

```python
food_shops = atlanta_poi.loc[(atlanta_poi['shop']=='bakery') | (atlanta_poi['shop']=='confectionery')\
                            | (atlanta_poi['shop']=='deli') | (atlanta_poi['shop']=='frozen yogurt')\
                            | (atlanta_poi['shop']=='Frozen Yogurt')| (atlanta_poi['shop']=='ice_cream')| (atlanta_poi['shop']=='fresh juices & smoothies')]
food_shops
```

```python
#3: shopping - accesories, cosmetics, department_store, mall, skateshop, electronics, craft, gift, glam shop, household, interior_decoration
shopping_df = atlanta_poi.loc[(atlanta_poi['shop']=='accesories') | (atlanta_poi['shop']=='cosmetics')\
                            | (atlanta_poi['shop']=='department_store') | (atlanta_poi['shop']=='mall')\
                            | (atlanta_poi['shop']=='skateshop')| (atlanta_poi['shop']=='electronics')| (atlanta_poi['shop']=='craft')|(atlanta_poi['shop']=='gift')| (atlanta_poi['shop']=='glam shop')| (atlanta_poi['shop']=='household')|(atlanta_poi['shop']=='interior decoration')]
shopping_df
                              
```
